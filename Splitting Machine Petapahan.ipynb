```python
# splitting machine
import pandas as pd
import openpyxl
import numpy as np


file_path = "Split.xlsx"


workbook = openpyxl.load_workbook(file_path)
res = workbook["Res (K x H)"]
poc = workbook["Production & Open Completion"]
split = workbook["Split"]

excel_data = pd.read_excel(file_path, sheet_name="Split")

row_split_results = []
for row_split in split.iter_rows(min_row=3):
    date = row_split[0].value
    uwi = row_split[6].value
    row_split_results.append([pd.Timestamp(date), uwi])

row_poc_results = []
for row_poc in poc.iter_rows(min_row=2, min_col=4):
    date = row_poc[0].value
    uwi = row_poc[6].value
    row_poc_results.append([pd.Timestamp(date), uwi])
    

indexer = 1
for i in row_split_results:
    indexer += 1
    for j in row_poc_results:
        if i == j:
            for row in poc.iter_rows(min_row=indexer, max_row=indexer, min_col=17):
                for k in enumerate(row):
                    cell = poc.cell(row=indexer, column=k[0] + 17).value
                    if cell == "P":
                        header = poc.cell(row=1, column=k[0] + 17).value
                        for row_res in res.iter_rows(min_row=1):
                            if row_res[0].value == header:
                                for l in res.iter_rows(min_row=1, max_row=1, min_col=2):
                                    for m in enumerate(l):
                                        cell_res = res.cell(row=1, column=m[0] + 2).value
                                        if cell_res == i[1]:
                                            multiplier_temp = res.cell(row=row_res[0].row, column=m[0] + 2).value
                                            totalSum = 0
                                            for col_res in res.iter_rows(min_row=2, max_row=50, min_col=m[0] + 2, max_col=m[0] + 2):
                                                if col_res[0].value is not None:
                                                    totalSum += col_res[0].value
                                                else:
                                                    continue
                                            multiplier = multiplier_temp / totalSum
                                            print(multiplier)
                    else:
                        continue
        else:
            continue
    print("break")
```
